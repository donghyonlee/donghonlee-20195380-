```
class Scaler:
  def __init__(self, x_range=(-120, 14870), y_range=(-120, 14980)):
    x_range = np.array(x_range).astype('float64').reshape(-1, 1)
    y_range = np.array(y_range).astype('float64').reshape(-1, 1)
    self.scaler_x = MinMaxScaler(feature_range=(0, 512))
    self.scaler_x.fit(x_range)
    self.scaler_y = MinMaxScaler(feature_range=(0, 512))
    self.scaler_y.fit(y_range)

  def transform(self, x, y):
    x = np.array(x)
    y = np.array(y)
    x_trans = self.scaler_x.transform(x.reshape(-1, 1))
    y_trans = self.scaler_y.transform(y.reshape(-1, 1))

    x_trans = np.squeeze(x_trans)
    y_trans = np.squeeze(y_trans)
    return x_trans, 512 - y_trans
```

```
def refine_event(events, gameid, win_team):
  result = []
  lane_table = {'TOP_LANE': 0, 'MID_LANE': 1, 'BOT_LANE': 2}
  scaler = Scaler()
  for event in events:

    if 'killerId' not in event:
      continue

    # 0 for blue, 1 for red, 2 for minion
    if 1 <= event['killerId'] <= 5:
      killer_team = 0
    elif 6 <= event['killerId'] <= 10:
      killer_team = 1
    else:
      killer_team = 2

    if killer_team != win_team:
      continue

    if event['type'] == 'CHAMPION_KILL':
      x, y = event['position'].values()
      x, y = scaler.transform(x, y)
      lane = getLane(x, y)
      tmp = (gameid, event['type'], lane, killer_team, event['timestamp'])

    elif event['type'] == 'ELITE_MONSTER_KILL':
      tmp = (gameid, event['type'], event['monsterType'], killer_team, event['timestamp'])

    elif event['type'] == 'BUILDING_KILL':
      tmp = (gameid, event['type'], lane_table[event['laneType']], killer_team, event['timestamp'])

    else:
      continue

    result.append(tmp)
  return result
```

```
events = []
for match in matches:
  win_table = {'Win': 0, 'Fail': 1}
  gameid = match
  req = requests.get(match_url.format(match))
  win = win_table[ requests.get(match_url2.format(match)).json()['teams'][0]['win'] ]
  for frame in req.json()['frames']:
    if not frame['events']:
      continue
    frame_events = refine_event(frame['events'], gameid, win)
    events += frame_events

  time.sleep(8)
```
